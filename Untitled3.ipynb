```python
le = LabelEncoder()
y_train = le.fit_transform(y_train)
xg=XGBClassifier('binary:logistic')
xg.fit(x_train,y_train)
```

```python
print("Accuracy of test dataset:",xg.score(x_test,y_test)*100)
```
