```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from datetime import datetime
```

```python
df=pd.read_csv("/content/train.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.region_category.mode()[0]
```

```python
df["region_category"]=df["region_category"].fillna(df["region_category"].mode()[0])
```

```python
df["preferred_offer_types"]=df["preferred_offer_types"].fillna(df["preferred_offer_types"].mode()[0])
```

```python
df["points_in_wallet"]=df["points_in_wallet"].fillna(df["points_in_wallet"].mean())
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df["Gender"]=LabelEncoder().fit_transform(df['gender'])
df.drop(["gender"],axis=1)
```

```python
df=df.drop([9017],axis=0)
```

```python
df.isnull().sum()
```

```python
df["Region"]=LabelEncoder().fit_transform(df['region_category'])
df["Membership"]=LabelEncoder().fit_transform(df['membership_category'])
df["preference"]=LabelEncoder().fit_transform(df["preferred_offer_types"])
df["reference"]=LabelEncoder().fit_transform(df["joined_through_referral"])
df["medium"]=LabelEncoder().fit_transform(df["medium_of_operation"])
df["internet"]=LabelEncoder().fit_transform(df["internet_option"])
df["discount"]=LabelEncoder().fit_transform(df["used_special_discount"])
df["complaint"]=LabelEncoder().fit_transform(df["past_complaint"])
df["Feedback"]=LabelEncoder().fit_transform(df["feedback"])
```

```python
train=df.drop(['gender','region_category','membership_category',"preferred_offer_types","joined_through_referral","medium_of_operation","internet_option","used_special_discount","past_complaint","feedback"],axis=1)
```

```python
train.info()
```

```python
train=train.drop(['customer_id','Name','security_no','referral_id','avg_frequency_login_days','offer_application_preference','complaint_status'],axis=1)
```

```python
train.info()
```

```python
train['joining_date'] = pd.to_datetime(train['joining_date'])
```

```python
train['year']=train['joining_date'].dt.year
train['month']=train['joining_date'].dt.month
train['day']=train['joining_date'].dt.day
```

```python
train=train.drop(['joining_date'],axis=1)
```

```python
train.info()
```

```python
train['last_visit_time'] = pd.to_datetime(train['last_visit_time'])
```

```python
train['hour']=train['last_visit_time'].dt.hour
train['min']=train['last_visit_time'].dt.minute
train['sec']=train['last_visit_time'].dt.second
```

```python
train=train.drop(['last_visit_time'],axis=1)
```
